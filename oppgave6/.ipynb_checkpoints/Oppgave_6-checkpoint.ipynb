#### Sick absence:
I collect data on sickness absence 4th quarter 2021 in the municipalities of Troms and Finnmark from Statistisk Sentralbyrå,  
https://www.ssb.no/statbank/table/12448/

```python
from pyjstat import pyjstat
import requests
POST_URL = 'https://data.ssb.no/api/v0/no/table/12448/'
```

```python
payload = {
  "query": [
    {
      "code": "Region",
      "selection": {
        "filter": "agg_single:Komm2020",
        "values": ["5401", "5402", "5403", "5404", "5405", "5406", "5411", "5412", "5413", "5414", "5415", "5416", "5417", "5418", "5419", "5420", "5421",
          "5422", "5423", "5424", "5425", "5426", "5427", "5428", "5429", "5430", "5432", "5433", "5434","5435", "5436", "5437", "5438", "5439", "5440",
          "5441","5442", "5443", "5444"]
      }
    },
    {
      "code": "Kjonn",
      "selection": {
        "filter": "item",
        "values": [
          "0"
        ]
      }
    },
    {
      "code": "Alder",
      "selection": {
        "filter": "item",
        "values": [
          "00"
        ]
      }
    },
    {
      "code": "ContentsCode",
      "selection": {
        "filter": "item",
        "values": [
          "Sykefraversprosent"
        ]
      }
    },
    {
      "code": "Tid",
      "selection": {
        "filter": "item",
        "values": [
          "2021K4"
        ]
      }
    }
  ],
  "response": {
    "format": "json-stat2"
  }
}
```

```python
df = requests.post(POST_URL, json = payload)
df = pyjstat.Dataset.read(df.text)
SF_2021K4 = df.write('dataframe') 
SF_2021K4.head()
```

```python
# I rearrange the data and name the muncipalities 'Kommune' and the sickness absence 'SF'
SF_2021K4.rename(columns={'region': 'Kommune', 'value': 'SF'}, inplace=True)

SF_2021K4 = SF_2021K4[['Kommune', 'SF']] 
SF_2021K4.head()
```

I have now made a dataframe called SF-2021K4 that contains sickness absence for every muncipality in Troms and Finnmark in per cent 4th quarter 2021.

#### Population data:
I now retrieve population figures by gender for the same geographical area. Data are taken from Statistisk Sentralbyrå and are as of 1 January 2022,  
https://www.ssb.no/statbank/table/07459/


```python

from pyjstat import pyjstat
import requests
POST_URL = 'https://data.ssb.no/api/v0/no/table/07459/'
```

```python
payload = {
  "query": [
    {
      "code": "Region",
      "selection": {
        "filter": "agg:KommSummer",
        "values": ["K-5401", "K-5402", "K-5403", "K-5404","K-5405","K-5406", "K-5411", "K-5412", "K-5413", "K-5414", "K-5415", "K-5416", "K-5417", "K-5418",
          "K-5419","K-5420","K-5421","K-5422","K-5423","K-5424","K-5425","K-5426","K-5427","K-5428","K-5429","K-5430","K-5432","K-5433","K-5434","K-5435",
          "K-5436", "K-5437","K-5438", "K-5439", "K-5440", "K-5441", "K-5442", "K-5443", "K-5444" ]
      }
    },
    {
      "code": "Kjonn",
      "selection": {
        "filter": "item",
        "values": [
          "1",
          "2"
        ]
      }
    },
    {
      "code": "Alder",
      "selection": {
        "filter": "agg:Funksjonell1a",
        "values": [
          "F303"
        ]
      }
    },
    {
      "code": "Tid",
      "selection": {
        "filter": "item",
        "values": [
          "2022"
        ]
      }
    }
  ],
  "response": {
    "format": "json-stat2"
  }
}
```

```python
df_2 = requests.post(POST_URL, json = payload)
df_2 = pyjstat.Dataset.read(df_2.text)
BF = df_2.write('dataframe') 
print(BF)
```

I arrange the table so that it shows columns for gender and the sum of these.

```python
BF.rename(columns={'value': 'Befolkning', 'region': 'Kommune'}, inplace=True)
BF = BF[['Kommune', 'kjønn', 'Befolkning']]
BF_2021K4 = pd.pivot_table(BF, index = 'Kommune', columns = 'kjønn', values = 'Befolkning' )
BF_2021K4['Befolkning_totalt'] = BF_2021K4.sum(axis=1)
BF_2021K4.head()
```

I merge population data and sickness absence figures into one table and call it 'tot_tabl'.

```python
tot_tabl = BF_2021K4.merge(SF_2021K4, on='Kommune', how='left')
tot_tabl.head() 
```

```python
# I make a new variable 'størrelse' that is two times SF:
tot_tabl['størrelse'] = tot_tabl['SF'].apply(lambda SF: SF*2)
tot_tabl.head() 
```

I make a plot that shows the connection between sickness absence and population

```python
import numpy as np
from matplotlib import pyplot as plt

fig,ax=plt.subplots(figsize=(10,8))

#adding axis lables:
ax.set_ylabel('Population')
ax.set_xlabel('Sick absence in %')

#plotting the function:
ax.scatter(tot_tabl['SF'],tot_tabl['Befolkning_totalt'],  label='Sick absence in muncipality')
ax.legend(loc='upper right',frameon=False)
ax.set_title('Distributed sickness absence by municipality in Troms and Finnmark 4th quarter 2021', fontsize = 16, color='blue')

```

#### Interactive website:

I will now make an interactive version of this figure. I have used the code that was presented at lectures in the spring of 2022 by Espen Sirnes.

```python
from bokeh.plotting import figure

#creating figure:
p = figure(
        title = "Distributed sickness absence by municipality in Troms and Finnmark 4th quarter 2021", 
        x_axis_label = 'Sickness absence in %',
        y_axis_label = 'Population',
         
        tools="hover", 
        tooltips = [
        ("Muncipality","@Kommune"),
         ("Sick absence in %","@SF"),
         ("Population","@Befolkning_totalt"),
         ("Female","@Kvinner"),
         ("Male","@Menn")
            ],
    
        plot_height = 580,
        plot_width = 980)
```

```python
from bokeh.io import show, output_notebook
from bokeh.plotting import output_file

p.scatter(
    source= tot_tabl,
    x= 'SF', 
    y= 'Befolkning_totalt',
    
    size ='størrelse', 
    color = 'orange',
    alpha= 0.8,
    )
p.outline_line_alpha=0
p.grid[0].grid_line_alpha=0
p.grid[1].grid_line_alpha=0

#Creating the graph and saving as html
output_notebook()
output_file("Sykefravær_2021K4.html")
show(p)
```
